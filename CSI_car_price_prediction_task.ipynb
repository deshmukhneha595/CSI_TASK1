```python
sns.boxplot(df['Price'])
```

```python
plt.figure(figsize=(11, 6))
df['Manufacturer'].value_counts().plot(kind='bar')
plt.xlabel('Manufacturer')
plt.ylabel('Price')
plt.title('Counts of car Model Using Bar Graph')
plt.show()
```

```python
plt.figure(figsize=(9, 15))
top_Manufacturer = df['Manufacturer'].value_counts().nlargest(5)
top_Manufacturer.plot(kind='pie', autopct='%1.6f%%')
plt.title('Top 5 Manufacturers Distribution Using Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
hg=df['Price'].value_counts().nlargest(50)
# data['Price'].plot(kind='hist', bins=10)
hg.plot(kind='hist',edgecolor="black",bins=10)
plt.xlabel('Price')
plt.ylabel('Manufacturer')
plt.title('Manufacturer Distribution Using Histogram')
plt.show()
```

```python
sns.set(rc={'figure.figsize': (8, 6)})
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price', y='Color', data=df)
plt.title('Price vs. Color Using Scatter Plot')
plt.show()
```

```python
# Select the numerical columns for the heatmap
numeric_cols = ['ID', 'Price', 'Prod. year', 'Engine volume', 'Mileage', 'Doors', 'Airbags']
numeric_data = df[numeric_cols]

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
